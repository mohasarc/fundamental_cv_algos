```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def histogram(src_img):
    # flattenthe image to one dimensional array
    flat_im = src_img.flatten()
    hist, bin_edges = np.histogram(flat_im, bins=256)
    bin_mids = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.plot(bin_mids, hist)
    plt.title("histogram") 
    plt.show()
```

```python
im1 = cv2.imread('grayscale_1.jpg', cv2.IMREAD_GRAYSCALE)
histogram(im1)
```

```python
im2 = cv2.imread('grayscale_2.jpg', cv2.IMREAD_GRAYSCALE)
histogram(im2)
```
